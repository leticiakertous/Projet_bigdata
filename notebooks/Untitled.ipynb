```python
import re
import pandas as pd
from bs4 import BeautifulSoup

soup = BeautifulSoup(html, "html.parser")
elements = soup.find_all(True)

mots_types = [
    "Studio",
    "Colocation",
    "Logement en résidence",
    "Logement entier",
    "Logement T2",
    "Logement chez l'habitant"
]

data = []

for el in elements:
    txt = el.get_text(" ", strip=True)

    # garder seulement les blocs qui ressemblent à des annonces
    if "€" in txt and any(mot in txt for mot in mots_types) and len(txt) > 40:
        prixs = re.findall(r'\d+\s?€', txt)
        prix = prixs[0] if prixs else None

        type_logement = None
        for mot in mots_types:
            if mot in txt:
                type_logement = mot
                break

        data.append({
            "texte": txt,
            "prix": prix,
            "type_logement": type_logement,
            "source": "studapart"
        })

df2 = pd.DataFrame(data)

# enlever doublons
df2 = df2.drop_duplicates(subset=["texte", "prix"]).reset_index(drop=True)

print(df2.head(20))
print("Nombre de lignes après filtrage :", len(df2))
```

```python
import pandas as pd
import re

# on part de df2 ou df3 selon ton nom
df_clean = df2.copy()

# mots/phrases à exclure
mots_a_exclure = [
    "Accueil >",
    "Recherche en cours",
    "Pas de résultats",
    "Comment fonctionne",
    "Questions fréquentes",
    "Guide propriétaires",
    "Bon à savoir",
    "Budget moyen",
    "Vous êtes propriétaire",
    "FAQ",
    "Toutes nos",
    "Louer Logement étudiant"
]

for mot in mots_a_exclure:
    df_clean = df_clean[~df_clean["texte"].str.contains(mot, case=False, na=False)]

# enlever les doublons exacts
df_clean = df_clean.drop_duplicates(subset=["texte", "prix", "type_logement"]).reset_index(drop=True)

print(df_clean.head(20))
print("Nombre de lignes après nettoyage :", len(df_clean))
```

```python
def extraire_titre(texte, type_logement):
    if pd.isna(texte):
        return None
    
    t = texte.replace("\n", " ").strip()
    
    # enlever le type au début si présent
    if type_logement and type_logement in t:
        t = t.split(type_logement, 1)[-1].strip()
    
    # couper avant le prix si présent
    t = re.split(r'\d+\s?€', t)[0].strip()
    
    return t[:120] if t else None

df_clean["titre"] = df_clean.apply(
    lambda row: extraire_titre(row["texte"], row["type_logement"]),
    axis=1
)

print(df_clean[["type_logement", "prix", "titre"]].head(20))
```

```python
df_clean["prix_num"] = (
    df_clean["prix"]
    .str.replace("€", "", regex=False)
    .str.replace(" ", "", regex=False)
)

df_clean["prix_num"] = pd.to_numeric(df_clean["prix_num"], errors="coerce")

print(df_clean[["prix", "prix_num"]].head(10))
```

```python
df_final = df_clean.drop_duplicates(
    subset=["prix_num", "titre"]
).reset_index(drop=True)
df_final = df_final[df_final["prix_num"] > 200]
df_final = df_final[~df_final["texte"].str.contains("Colocations à Paris uniquement", na=False)]
df_final = df_final.dropna(subset=["prix_num", "titre"])
print(df_final.head(10))
print("Nombre après suppression doublons :", len(df_final))
```

```python
import re

def extract_surface(text):
    match = re.search(r'(\d+)\s?m2', text.lower())
    return int(match.group(1)) if match else None

df_final["surface"] = df_final["texte"].apply(extract_surface)
```

```python
def extract_pieces(text):
    match = re.search(r'(\d+)\s?(chambres?|pièces?)', text.lower())
    return int(match.group(1)) if match else None

df_final["nb_pieces"] = df_final["texte"].apply(extract_pieces)
```

```python
def extract_ville(text):
    if "paris" in text.lower():
        return "Paris"
    elif "colombes" in text.lower():
        return "Colombes"
    else:
        return None

df_final["ville"] = df_final["texte"].apply(extract_ville)
```

```python
df_final["description"] = df_final["texte"]
```

```python
df_model = df_final[
    ["prix_num", "type_logement", "surface", "nb_pieces", "ville", "description"]
]

print(df_model.head())
```
